```python
# 라이브러리 임포트
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score

# 데이터 불러오기
train = pd.read_csv("/content/train.csv")
test = pd.read_csv("/content/test.csv")

# X, y 분리
X = train.drop("BodyFat", axis=1)
y = train["BodyFat"]

# 결측치 처리
X = X.fillna(X.median())
test = test.fillna(test.median())

# 스케일링
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
test_scaled = scaler.transform(test)

# 학습/검증 데이터 분리
X_train, X_valid, y_train, y_valid = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# KNN 모델 (튜닝 버전)
model = KNeighborsRegressor(
    n_neighbors=5,           # 근접 이웃 수
    weights='distance',      # 거리 가중치 적용
    p=2                      # 유클리드 거리
)
model.fit(X_train, y_train)

# 검증 데이터 R² 확인
y_pred_valid = model.predict(X_valid)
r2 = r2_score(y_valid, y_pred_valid)
print(f"Validation R²: {r2:.5f}")

# 테스트 데이터 예측
test_pred = model.predict(test_scaled)

# 제출 파일 생성 
df = pd.read_csv("/content/submission.csv")
df.dropna(axis=1, inplace=True)  # 불필요한 컬럼 제거

# KNN 모델 예측값으로 대체
y_pred = test_pred

df["BodyFat"] = y_pred
df.to_csv("new_submission.csv", index=False)

print("new_submission.csv 파일이 생성되었습니다.")
```
